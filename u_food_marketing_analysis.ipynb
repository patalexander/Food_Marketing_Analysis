```python
age_groups = [(23,30),(31,40),(41,50),(51,60),(61,70),(71,85)]


def assign_age_group(Age):
    for age_range in age_groups:
        if age_range[0] <= Age <= age_range[1]:
            return f"{age_range[0]} - {age_range[1]}"
    return('Unknown')

food['Age_Group'] = food['Age'].apply(assign_age_group)
        
  
```

### 2. Catalog was more likely to accept campaigns, but in person spend more. Recommend a split between all -    (40% catalog, 30% store, 30% web)
### 3. Focus on people with no kids (or less kids).
### 4. Education has no impact - don't target any group
### 5. Marital status does not play a big part.  Married, Single, Together spend more money
